## Generation of random data points with normal distribution

```python
# Parameters of the centroids distributions
I=identity(2) #Covariance matrix
Mb=[1,0] #Blue mean
Mo=[0,1] #Orange mean
s=10 #Number of centroids
```

```python
#Generation of centroids
bk=random.multivariate_normal(Mb,I,s) #Blue Centroids
ok=random.multivariate_normal(Mo,I,s) #Orange Centroids
```

```python
# Visualization of centroids
figure(figsize=(6,6))
scatter(bk[:,0],bk[:,1],marker='o',s=50,c='red')
scatter(ok[:,0],ok[:,1],marker='o',s=50,c='blue')
show()
```

```python
# Function that creates random numbers from normal distribution with normal mean
# It gets for parameters the number of points, the array of means to choose randomly,
# and the covariance matrix S
# It returns a 2xn matrix of x and y data

def gen_random_samples(n,means,S):
    samples=zeros([2,n]) # Initialize matrix with correct dimensions
    for i in range(n):
        r=randint(0,len(means)-1) #Chooses a random mean
        m=[float(means[r,0]),float(means[r,1])]
        samples[:,i]=random.multivariate_normal(m,S,1) #Fill matrix
    return samples
```

```python
# Generate blue (B) and orange (O) points
B=gen_random_samples(200,bk,I/5)
O=gen_random_samples(200,ok,I/5)
```

```python
# Visualize points
figure(figsize=(12,12))
scatter(B[0,:],B[1,:],marker='o',s=50,c='red')
scatter(O[0,:],O[1,:],marker='o',s=50,c='blue')
show()
```

## Linear Regression:<BR>  Calculates the $\beta$ vector with the parameters for the hyperplane that divides the two sets of data

Using equation (2.5) of the textbook *Elements of Statistical Learning*, we can find the vector $\beta$ with the coefficients describing the hyperplane:

$$ \hat{\beta}=(\textbf{X}^\intercal\textbf{X})^{-1}\textbf{X}^\intercal\textbf{y} $$

The $n\times1$ vector $\textbf{y}$ is the output of the classification. This output is quantified turning cathegorical variables into numbers. In this case blue is 0 and orange is 1.

```python
# Create the X matrix with blue and orange data
X=append(transpose(B),transpose(O),axis=0)
# Create output (1 and 0)
yb=zeros([len(B[0,:]),1])
yo=ones([len(O[0,:]),1])
Y=append(yb,yo,axis=0)
# Create transpose of X
Xt=transpose(X)
# Calculate beta with equation (2.5)
Beta=dot( dot( inv(dot(Xt,X)) , Xt ) , Y)
# Specify slope and intersection
b=Beta[0]
m=Beta[1]
```

```python
# Set limits for the plot
Xmax=ceil(max(max(B[0,:]),max(O[:,0]))+1)
Ymax=ceil(max(max(B[1,:]),max(O[:,1]))+1)
Xmin=floor(min(min(B[0,:]),min(O[:,0]))-1)
Ymin=floor(min(min(B[1,:]),min(O[:,1]))-1)
```

```python
# Define the equation of the hyperplane
def division(x,m,b):
    return m*x+b
```

```python
# Create coordinate maps for the plot
x=arange(Xmin,Xmax,0.01)
y=arange(Ymin,Ymax,0.01)
xx,yy=meshgrid(x,y)
# Establish the regions of classification in the plot
z=zeros(xx.shape)
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        if yy[i,j]<division(xx[i,j],m,b):
            z[i,j]=1
```

```python
# Visualize linear classification
figure(figsize(12,12))
cm = plt.cm.RdBu
cs=contourf(xx,yy,-z,cmap=cm, alpha=0.25)
plot(x,division(x,m,b),'k',linewidth=2)
scatter(B[0,:],B[1,:],marker='o',s=50,facecolors='none', edgecolors='r')
scatter(O[0,:],O[1,:],marker='o',s=50,facecolors='none', edgecolors='b')
title('Linear Regression of 0/1 Response')
show()
```

```python
# Error
def calc_error(B,O):
    cont=0
    for i in range(len(B[0,:])):
        if B[1,i]>m*B[0,i]+b:
            cont+=1   
    for j in range(len(O[0,:])):
        if O[1,j]<m*O[0,j]+b:
            cont+=1
    return cont/float(len(B[0,:])+len(O[0,:]))
```

## Nearest-Neighbor

```python
from sklearn import neighbors
```

### Create a k-neighbour model with k=15

```python
model15 = neighbors.KNeighborsClassifier(15, weights='distance')
model15.fit(X, Y)
# Create the colormap
Z = model15.predict(c_[xx.ravel(), yy.ravel()])
# Reshape the answer to fit the coordinates
Z=Z.reshape(xx.shape)
```

```python
# Visualize k-neighbour classification with k=15
figure(figsize(12,12))
cm = plt.cm.RdBu
cs=contourf(xx,yy,Z,cmap=cm, alpha=0.25)
cs2=contour(xx,yy,Z,color='k',lw=3)
scatter(B[0,:],B[1,:],marker='o',s=50,facecolors='none', edgecolors='r')
scatter(O[0,:],O[1,:],marker='o',s=50,facecolors='none', edgecolors='b')
title('15-Nearest Neighbor Classifier')
show()
```

### Create a k-neighbour model with k=1

```python
model1 = neighbors.KNeighborsClassifier(1, weights='distance')
model1.fit(X, Y)
# Create the colormap
Z = model1.predict(c_[xx.ravel(), yy.ravel()])
# Reshape the answer to fit the coordinates
Z=Z.reshape(xx.shape)
```

```python
# Visualize k-neighbour classification with k=1
figure(figsize(12,12))
cm = plt.cm.RdBu
cs=contourf(xx,yy,Z,cmap=cm, alpha=0.25)
cs2=contour(xx,yy,Z,color='k',lw=3)
scatter(B[0,:],B[1,:],marker='o',s=50,facecolors='none', edgecolors='r')
scatter(O[0,:],O[1,:],marker='o',s=50,facecolors='none', edgecolors='b')
title('1-Nearest Neighbor Classifier')
show()
```

```python
def calc_kerror(B,O,Z):
    cont=0
    for i in range(len(B[0,:])):
        if Z[B[0,i],B[1,i]]!=0:
            cont+=1
    for j in range(len(O[0,:])):
        if Z[O[0,j],O[1,j]]!=1:
            cont+=1
    return cont/float(len(B[0,:])+len(O[0,:]))
```
